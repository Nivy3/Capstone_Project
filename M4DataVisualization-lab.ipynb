# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plt.hist(x='ConvertedComp', data=df)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY= """
SELECT Age 
FROM master
"""
df1 = pd.read_sql_query(QUERY, conn)
sns.boxplot(data=df1)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
dfs = pd.read_sql_query(QUERY,conn)
plot= sns.scatterplot(x='Age', y='WorkWeekHrs', data=dfs, color = 'red')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
dfb=pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=dfb, x='WorkWeekHrs', y='CodeRevHrs',size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age', size=14) 
plt.xlabel('WorkWeekHrs', size=10) 
plt.ylabel('CodeRevHrs', size=10)

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

dfp=pd.read_sql_query(QUERY,conn)  
dfp.set_index('DatabaseDesireNextYear', inplace=True) 
print(dfp)

dfp['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', startangle=90, shadow=True,) 

plt.legend(labels=dfp.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
```

```python
QUERY = """ 
SELECT LanguageDesireNextYear, count(LanguageDesireNextYear) as Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
order by count(LanguageDesireNextYear) DESC LIMIT 5 
""" 

dfp1=pd.read_sql_query(QUERY,conn)  
dfp1.set_index('LanguageDesireNextYear', inplace=True) 
print(dfp1)

dfp1['Count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', startangle=90, shadow=True,) 

plt.legend(labels=dfp1.index, loc='upper right')
plt.title('Most polpular languages people wish to learn next year') 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY,conn)
df_s = df.groupby('Age').median()

df_s.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY,conn)
df_l = df.groupby('Age').median()

df_l.plot(kind='line', figsize=(20, 6))

plt.title('Median ConvertedComp for Those Age 45 to 60')
plt.ylabel('ConvertedComp')
plt.show()
```

```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 25 AND 30
"""
df = pd.read_sql_query(QUERY,conn)
df_ls = df.groupby('Age').median()

df_ls.plot(kind='line', figsize=(20, 6))

plt.title('Median ConvertedComp for Those Age 25 to 30')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranch
from master
group by MainBranch
"""

df_b=pd.read_sql_query(QUERY,conn)

df_b.plot(kind='barh', figsize=(10,6), color='orange')

plt.show()
```
